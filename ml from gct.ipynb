```python
import pandas as pd
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,7,8,9]
y = [1,2,3,4,5,6,7,8,9]

data = {'x':x, 'y':y}
df = pd.DataFrame(data)
print(df)

plt.scatter(df['x'],df['y'])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('scatter Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy import stats

x = [-1,-2,1,2,3,4,5,6,7,8,9]
y = [-1,-2,1,2,3,4,5,6,7,8,9]


slope,intercept,r,p,std_err = stats.linregress(x,y)

def myfunc(x):
    return slope * x + intercept

mymodel = list(map(myfunc,x))

plt.scatter(x,y)
plt.plot(x, mymodel)
plt.show()
```
